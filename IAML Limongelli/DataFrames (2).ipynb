Creare un dataFrame

```python
import pandas as pd

dati={'Nome':['Luca','Sara','Marco'],
      'Età':[25,30,22],
      'Città':['Roma','Torino','Napoli']}
df=pd.DataFrame(dati)
print(df)
print("-"*50)
print("{:.2f}".format(df['Età'].mean()))
```

Accedere ai dati

```python
print(df['Nome'])  #stampa tutti i dati della colonna nome
```

```python
print(df.iloc[0]) #stampa la riga di indice 0, trasponendo righe e colonne
print(df)
```

```python
#stampa il dato che si trova nell'indice di riga i e nel campo etichettato
print(df.loc[1,'Città'])
```

# Filtro e pulizia dei dati

```python
dn = pd.read_csv('sample_data/studentiMancanti.csv')
print(dn)
```

Se voglio eliminare le righe che contengono dati mancanti

```python
print(dn.dropna()) #se metto inplace=True modifica il dataframe originale
print("-"*50)
print(dn)
```

```python
# il data frame non viene modificato dai metodi,
#se si vuole modificare un dataframe bisogna crearne uno nuovo
dn_pulita=dn.dropna()
print(dn_pulita)
print('-'*30)
print(dn)
```

Se voglio trasporre la tabella

```python
print(dn.T)
```

Per sostituire NaN con 0

```python
print(dn.fillna(0))
print('-'*50)
print(dn)
```

Per sostituire Nan con valori particolari es: con 'mancante' o con la media delle età

```python
print(dn.fillna({'Nome': 'mancante', 'Età': df['Età'].mean()}))
```

MERGE DI DATAFRAMES

```python
df1 = pd.DataFrame({'ID': [1,2], 'Nome': ['Luca','Sara']})
df2 = pd.DataFrame({'ID': [1,2], 'Voto': [90,85]})
print(df1)
print('-'*50)
print(df2)
```

```python
print(pd.merge(df1, df2, on='ID'))
```

```python
pd.concat([df1, df2], axis=0)
```
